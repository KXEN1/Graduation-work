```python
import cv2
import easyocr
import json
from PIL import Image, ImageDraw
import matplotlib.pyplot as plt

# EasyOCR reader   (   )
reader = easyocr.Reader(['ko', 'en'], gpu=False)

def extract_text_from_image(image):
    """
         .
    """
    #   
    results = reader.readtext(image)
    
    #     PIL  
    img = Image.fromarray(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    draw = ImageDraw.Draw(img)

    #    
    extracted_text = []
    
    #       
    for detection in results:
        bbox = detection[0]
        text = detection[1]
        extracted_text.append(text)
        
        #  
        x_min, y_min = int(bbox[0][0]), int(bbox[0][1])
        x_max, y_max = int(bbox[2][0]), int(bbox[2][1])
        
        #  
        draw.rectangle([x_min, y_min, x_max, y_max], outline="blue", width=2)
        draw.text((x_min, y_min - 20), text, fill="blue")
    
    #  
    plt.figure(figsize=(25, 18))
    plt.imshow(img)
    plt.axis('off')
    plt.show()
    
    return extracted_text

def save_text_to_json(text_list, file_path):
    """
       JSON  .
    """
    data = {
        "Extracted Text": text_list
    }
    
    with open(file_path, 'w', encoding='utf-8') as json_file:
        json.dump(data, json_file, ensure_ascii=False, indent=4)
    print(f"Extracted text has been saved to {file_path}")

#   
image_path = 'bill2.jpeg'

#  
image = cv2.imread(image_path)
if image is None:
    raise Exception(f"Could not read the image from {image_path}")

#     
extracted_text = extract_text_from_image(image)

# JSON  
json_file_path = 'extracted_text.json'
save_text_to_json(extracted_text, json_file_path)
```

```python
import re
import json

def load_text_from_json(json_file_path):
    """
    JSON    
    """
    with open(json_file_path, 'r', encoding='utf-8') as json_file:
        data = json.load(json_file)
        return data.get("Extracted Text", [])

def extract_business_numbers(text_list):
    """
        
    """
    business_number_pattern = re.compile(r'\b\d{3}-\d{2}-\d{5}\b')
    business_numbers = []

    for text in text_list:
        matches = business_number_pattern.findall(text)
        business_numbers.extend(matches)

    return business_numbers

def extract_prices(text_list):
    """
            
    """
    price_pattern = re.compile(r'.*?([$]?\s?\d{1,3}(?:,\d{3})*(?:\.\d{2})?)')
    prices = []

    for text in text_list:
        matches = price_pattern.findall(text)
        prices.extend(matches)

    return prices

def extract_store_names(text_list):
    """
       
    """
    store_name_pattern = re.compile(r'(||||)\s*[:]?\s*(\S+)')
    store_names = []

    for text in text_list:
        matches = store_name_pattern.findall(text)
        if matches:
            for match in matches:
                store_names.append(match[1])

    return store_names

def extract_transaction_date(text_list):
    """
      , ,        
    """
    date_pattern = re.compile(r'(||||)\s*[:]?\s*(\d{4}[-/]\d{2}[-/]\d{2})\s+(\d{2}):\s?(\d{2}):\s?(\d{2})')
    dates = []

    for text in text_list:
        matches = date_pattern.findall(text)
        for match in matches:
            full_date = f"{match[1]} {match[2]}:{match[3]}:{match[4]}"
            dates.append(full_date)

    return dates

def extract_item_prices(text_list):
    """
        
    """
    item_price_pattern = re.compile(r'(||)\s*[:]?\s*(.*?)\s*\s*[:]?\s*([$]?\s?\d{1,3}(?:,\d{3})*(?:\.\d{2})?)')
    item_prices = []

    for text in text_list:
        matches = item_price_pattern.findall(text)
        for match in matches:
            item_prices.append((match[1], match[2]))  # (, )

    return item_prices

def extract_total_price(text_list):
    """
        
    """
    total_price_pattern = re.compile(r'\s*\s*[:]?\s*([$]?\s?\d{1,3}(?:,\d{3})*(?:\.\d{2})?)')
    total_prices = []

    for text in text_list:
        matches = total_price_pattern.findall(text)
        total_prices.extend(matches)

    return total_prices

# JSON  
json_file_path = 'extracted_text.json'

# JSON   
extracted_text = load_text_from_json(json_file_path)

#   
business_numbers = extract_business_numbers(extracted_text)
prices = extract_prices(extracted_text)
store_names = extract_store_names(extracted_text)
transaction_dates = extract_transaction_date(extracted_text)
item_prices = extract_item_prices(extracted_text)
total_prices = extract_total_price(extracted_text)

#   
print(" :", business_numbers)
print(":", prices)
print(":", store_names)
print(" :", transaction_dates)
print(" :", item_prices)
print(" :", total_prices)
```

```python
import json
from selenium import webdriver
from selenium.webdriver.common.by import By

def extract_category_keywords(business_numbers):
    """
               .
    
    Args:
        business_numbers (list):   .
    
    Returns:
        dict:        .
    """
    # Selenium driver 
    driver = webdriver.Chrome()
    
    #    
    category_keywords_dict = {}

    #        
    for business_number in business_numbers:
        #    "-"  (URL    )
        business_number_clean = business_number.replace("-", "")
        
        #  URL 
        address = 'https://bizno.net/article/' + business_number_clean
        print(f"  URL: {address}")

        # URL 
        driver.get(address)

        # category_keywords 
        try:
            category_keywords = driver.find_element(By.XPATH, '/html/body/section[2]/div/div/div[1]/div[1]/div/table/tbody/tr[4]/td/p[3]').text
            print(f"  {business_number}  : {category_keywords}")
            category_keywords_dict[business_number] = category_keywords
        except Exception as e:
            print(f"  {business_number}     . : {e}")
            category_keywords_dict[business_number] = None

    #  
    driver.quit()

    #  
    return category_keywords_dict

def save_to_json(data, file_path):
    """
     JSON   .
    
    Args:
        data (dict):  .
        file_path (str):  JSON  .
    """
    with open(file_path, 'w', encoding='utf-8') as json_file:
        json.dump(data, json_file, ensure_ascii=False, indent=4)
    print(f" {file_path} JSON  .")
    

#     
category_keywords = extract_category_keywords(business_numbers)

#      JSON  
if category_keywords:
    json_file_path = 'category_keywords.json'
    save_to_json(category_keywords, json_file_path)
else:
    print("    JSON   .")
```
